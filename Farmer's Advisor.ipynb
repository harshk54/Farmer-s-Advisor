```python
def crop_data(state,district,season,area):
    import pandas as pd  
    import numpy as np 
    df = pd.read_csv('Desktop/data.csv') 
    list_maize=[]
    list_rice=[]
    list_wheat=[]
    list_year=[]
    result={}


    for i in range(len(df.index)):
        if(df.loc[i].State.lower()==state.lower() and df.loc[i].District.lower()==district.lower() and df.loc[i].Season.lower()==season.lower()):
            if(df.loc[i].Crop=="Maize"):
                list_maize.append(df.loc[i].Yield)
            elif(df.loc[i].Crop=="Wheat"):
                list_wheat.append(df.loc[i].Yield)
            elif(df.loc[i].Crop=="Rice"):
                list_rice.append(df.loc[i].Yield)


    for i in range(2010,2020):
        list_year.append(i)

    list_year

    df_temp=pd.DataFrame({'Year':list_year})



    ## Maize Estimation

    df_maize=df_temp
    df_maize["Yield"]=list_maize

    X = np.array(df_maize["Year"]).reshape(-1, 1) 
    y = np.array(df_maize["Yield"]).reshape(-1, 1) 

    from sklearn.linear_model import LinearRegression

    classifier = LinearRegression()  
    classifier.fit(X, y)

    y_test=[2020]

    df_test=pd.DataFrame({'test':y_test})
    y_pred = classifier.predict(df_test)  

    result["Maize"]=y_pred[0][0]



    ## Wheat Estimation

    df_wheat=df_temp
    df_wheat["Yield"]=list_wheat

    X = np.array(df_wheat["Year"]).reshape(-1, 1) 
    y = np.array(df_wheat["Yield"]).reshape(-1, 1) 

    from sklearn.linear_model import LinearRegression

    classifier = LinearRegression()  
    classifier.fit(X, y)

    y_test=[2020]

    df_test=pd.DataFrame({'test':y_test})
    y_pred = classifier.predict(df_test)  
    result["Wheat"]=y_pred[0][0]

    ## Rice Estimation

    df_rice=df_temp
    df_rice["Yield"]=list_rice

    X = np.array(df_rice["Year"]).reshape(-1, 1) 
    y = np.array(df_rice["Yield"]).reshape(-1, 1) 

    from sklearn.linear_model import LinearRegression

    classifier = LinearRegression()  
    classifier.fit(X, y)

    y_test=[2020]

    df_test=pd.DataFrame({'test':y_test})
    y_pred = classifier.predict(df_test)  
    result["Rice"]=y_pred[0][0]
    
    return result
    
result1=crop_data("Tamil Nadu","Vellore","Zaid",1000)
result2=crop_data("Tamil Nadu","Chennai","Zaid",1000)
result3=crop_data("Tamil Nadu","Coimbatore","Zaid",1000)
```
